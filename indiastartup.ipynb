```python

# Handling missing values in 'Day', 'Year', and 'Month' columns
df['Day'] = df['Day'].apply(lambda x: pd.NA if x in ['CO', 'Tuesday'] else x)
df['Year'] = df['Year'].replace({'NaN': pd.NA})
df['Month'] = df['Month'].replace({'NaN': pd.NA})

# Impute missing 'Day' values with mode (most common day)
df['Day'] = df['Day'].fillna(df['Day'].mode()[0])

# Impute missing 'Year' values with median
df['Year'] = df['Year'].fillna(df['Year'].median())

# Impute missing 'Month' values with mode (most common month)
df['Month'] = df['Month'].fillna(df['Month'].mode()[0])

# Print the updated DataFrame with imputed values
print(df)
```

```python
#Calculate the total amount invested in each industry vertical.

industry_investments = df.groupby('IndustryVertical')['AmountInUSD'].value_counts().head(30)
print('The total amount invested in each industryvertical are:')
print(industry_investments)
```

```python
#Plot the distribution of investments by industry vertical
plt.figure(figsize=(12, 6))
industry_investments.plot(kind='bar', color='skyblue')
plt.title('Distribution of Investments by Industry Vertical')
plt.xlabel('Industry Vertical')
plt.ylabel('Total Amount Invested (in USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Remove rows with missing or invalid values in 'AmountInUSD' column
df = df[df['AmountInUSD'] != '-']

# Convert 'AmountInUSD' column to numeric type
df['AmountInUSD'] = pd.to_numeric(df['AmountInUSD'], errors='coerce')

# Calculate the total amount invested in each city
city_investments = df.groupby('City')['AmountInUSD'].sum().sort_values(ascending=False)
print('The total amount invested in each city are:')
print(city_investments)


plt.figure(figsize=(12, 6))
city_investments.plot(kind='bar', color='skyblue')
plt.title('Geographical Distribution of Investments by City')
plt.xlabel('City')
plt.ylabel('Total Amount Invested (in USD)')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(12, 6))
df['IndustryVertical'].value_counts().plot(kind='bar')
plt.xlabel('Industry Vertical')
plt.ylabel('Number of Startups')
plt.title('Count of Startups in Each Industry Vertical')
plt.show()
```
